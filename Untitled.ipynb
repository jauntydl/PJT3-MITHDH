```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from keras.utils import to_categorical

datapath = 'resources/globalsales.csv'
df = pd.read_csv(datapath,encoding = "ISO-8859-1")
df.info()
```

```python
df.drop(df.columns[[0,1,2,3,5,6,7,8,9,10,11,12,13,16,17,25]], axis=1, inplace=True)
df.columns =['date',"region","market","sub-cat","cat","sales","quantity","discount","profit","shippingcost"]
df['date'] = df['date'].apply(lambda x: pd.to_datetime(str(x), format='%Y-%m-%d'))
```

```python
df.info()
```

```python
df
```

```python
df_main = df.groupby(['date','region','market','sub-cat','cat'], as_index = False)['sales','profit','shippingcost'].sum()
```

```python
df_main
```

```python
df_main.info()
```

```python
y = df_main.iloc[:,1]
label_encoder1 = LabelEncoder()
label_encoder1.fit(y)
encoded_y = label_encoder1.transform(y)
region_en = to_categorical(encoded_y)

y = df_main.iloc[:,2]
label_encoder2 = LabelEncoder()
label_encoder2.fit(y)
encoded_y = label_encoder2.transform(y)
market_en = to_categorical(encoded_y)

y = df_main.iloc[:,3]
label_encoder3 = LabelEncoder()
label_encoder3.fit(y)
encoded_y = label_encoder3.transform(y)
subcat_en = to_categorical(encoded_y)

y = df_main.iloc[:,4]
label_encoder4 = LabelEncoder()
label_encoder4.fit(y)
encoded_y = label_encoder4.transform(y)
cat_en = to_categorical(encoded_y)
```
