```python
#create a new dataframe to model the difference
df_diff = test_df.copy()
#add previous sales to the next row
df_diff['prev_sales'] = df_diff['sales'].shift(1)
df_diff['date'] = df_diff.index
#drop the null values and calculate the difference
df_diff['diff'] = (df_diff['sales'] - df_diff['prev_sales'])
df_diff.head(10)
```

```python
plt.plot(df_diff['diff'])
```

```python
#create dataframe for transformation from time series to supervised
df_supervised = df_diff.drop(['prev_sales'],axis=1)
#adding lags
for inc in range(1,13):
    field_name = 'lag_' + str(inc)
    df_supervised[field_name] = df_supervised['diff'].shift(inc)
#drop null values
df_supervised = df_supervised.dropna().reset_index(drop=True)
```

```python
df_supervised
```

```python
# Import statsmodels.formula.api
import statsmodels.formula.api as smf
# Define the regression formula
model = smf.ols(formula='diff ~ lag_1 + lag_2 +lag_3 +lag_4 +lag_5 +lag_6 +lag_7 +lag_8 +lag_9 +lag_10 +lag_11 +lag_12', data=df_supervised)
# Fit the regression
model_fit = model.fit()
# Extract the adjusted r-squared
regression_adj_rsq = model_fit.rsquared_adj
print(regression_adj_rsq)
```

```python
def get_stationarity(timeseries):
    
    # rolling statistics
    rolling_mean = timeseries.rolling(window=12).mean()
    rolling_std = timeseries.rolling(window=12).std()
    
    # rolling statistics plot
    original = plt.plot(timeseries, color='blue', label='Original')
    mean = plt.plot(rolling_mean, color='red', label='Rolling Mean')
    std = plt.plot(rolling_std, color='black', label='Rolling Std')
    plt.legend(loc='best')
    plt.title('Rolling Mean & Standard Deviation')
    plt.show(block=False)
    
    # Dickey–Fuller test:
    result = adfuller(timeseries)
    print('ADF Statistic: {}'.format(result[0]))
    print('p-value: {}'.format(result[1]))
    print('Critical Values:')
    for key, value in result[4].items():
        print('\t{}: {}'.format(key, value))
```

```python
def filterdata(df_i,result='sales', region = True, market = True, subcat = True, cat = True):
    temp = ['date','seq_date']
    if region == True:
        temp.append("region")
    if market == True:
        temp.append("market")
    if subcat == True:
        temp.append("sub-cat")
    if cat == True:
        temp.append("cat")
            
    df_filtered = df_i.groupby(temp, as_index = False)["{}".format(result)].sum()
    return df_filtered
```
