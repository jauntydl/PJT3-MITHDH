```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
from keras.utils import to_categorical

from statsmodels.tsa.arima_model import ARIMA
from sklearn.metrics import mean_squared_error
from matplotlib import pyplot as plt
from statsmodels.tsa.stattools import adfuller
from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.tsa.arima_model import ARIMA
from pandas.plotting import register_matplotlib_converters
register_matplotlib_converters()

def filterdata(df_i,result='sales', region = False, market = False, subcat = False, cat = False):
    temp = ['date']
    if region == True:
        temp.append("region")
    if market == True:
        temp.append("market")
    if subcat == True:
        temp.append("sub-cat")
    if cat == True:
        temp.append("cat")
            
    df_filtered = df_i.groupby(temp, as_index = False)["{}".format(result)].sum()
    return df_filtered

def to_integer(dt_time):
    return (dt_time.year-2016)*12 + dt_time.month


datapath = 'resources/globalsales.csv'
df = pd.read_csv(datapath,encoding = "ISO-8859-1")

df.drop(df.columns[[0,1,2,3,5,6,7,8,9,10,11,12,13,16,17,25]], axis=1, inplace=True)
df.columns =['date',"region","market","sub-cat","cat","sales","quantity","discount","profit","shippingcost"]
df['date'] = df['date'].apply(lambda x: pd.to_datetime(str(x), format='%Y-%m-%d'))
df['date'] = df['date'].apply(lambda dt: dt.replace(day=1))

df_main = df.groupby(['date','region','market','sub-cat','cat'], as_index = False)['sales','profit','shippingcost'].sum()


# seq_month = []

# for n in df_main['date']:
#     seq_month.append(to_integer(n))

# df_main["seq_month"] = seq_month

df_main_train = df_main[df_main["seq_month"] < 37]
df_main_val = df_main[df_main["seq_month"] >= 37]
```

```python
def get_stationarity(timeseries):
    
    # rolling statistics
    rolling_mean = timeseries.rolling(window=12).mean()
    rolling_std = timeseries.rolling(window=12).std()
    
    # rolling statistics plot
    original = plt.plot(timeseries, color='blue', label='Original')
    mean = plt.plot(rolling_mean, color='red', label='Rolling Mean')
    std = plt.plot(rolling_std, color='black', label='Rolling Std')
    plt.legend(loc='best')
    plt.title('Rolling Mean & Standard Deviation')
    plt.show(block=False)
    
    # Dickeyâ€“Fuller test:
    result = adfuller(timeseries)
    print('ADF Statistic: {}'.format(result[0]))
    print('p-value: {}'.format(result[1]))
    print('Critical Values:')
    for key, value in result[4].items():
        print('\t{}: {}'.format(key, value))
```

```python
get_stationarity(test_df)
```

```python
df_log = np.log(test_df['sales'])
plt.plot(df_log)
```

```python
df_log_shift = df_log - df_log.shift()
df_log_shift.dropna(inplace=True)
get_stationarity(df_log_shift)
```

```python
decomposition = seasonal_decompose(df_log) 
model = ARIMA(df_log, order=(2,1,2))
results = model.fit(disp=-1)
plt.plot(df_log_shift)
plt.plot(results.fittedvalues, color='red')
```
